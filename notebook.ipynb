### Cleaning Data

Untuk memudahkan pengambilan data, akan lebih efektif jika informasi yang disajikan dalam bentuk deskripsi yang lebih jelas. Saat ini, kondisi 'musim' dan 'cuaca' di tabel hanya ditampilkan sebagai angka yang merepresentasikan keadaan sebenarnya.

Oleh karena itu, kita sebaiknya mengonversinya menjadi keterangan yang lebih deskriptif, sehingga dapat lebih mudah dipahami.

````python
# Transformasi data
day_df['yr'] = day_df['yr'].map({0: '2011'})
day_df['mnth'] = day_df['mnth'].map({1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'Mei', 6: 'Jun'})
day_df['mnth'] = pd.Categorical(day_df['mnth'], categories=['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun'], ordered=True)
day_df['season'] = day_df['season'].map({
    1: 'Musim Semi', 2: 'Musim Panas', 3: 'Musim Gugur', 4: 'Musim Dingin'
})
day_df['weekday'] = day_df['weekday'].map({
    0: 'Minggu', 1: 'Senin', 2: 'Selasa', 3: 'Rabu', 4: 'Kamis', 5: 'Jumat', 6: 'Sabtu'
})
day_df['weathersit'] = day_df['weathersit'].map({
    1: 'Cerah', 2: 'Berawan', 3: 'Hujan/Salju Ringan', 4: 'Cuaca Buruk'
})
day_df['dteday'] = pd.to_datetime(day_df.dteday)
day_df['season'] = day_df.season.astype('category')
day_df['yr'] = day_df.yr.astype('category')
day_df['mnth'] = day_df.mnth.astype('category')
day_df['holiday'] = day_df.holiday.astype('category')
day_df['weekday'] = day_df.weekday.astype('category')
day_df['workingday'] = day_df.workingday.astype('category')
day_df['weathersit'] = day_df.weathersit.astype('category')

day_df.head()
````

Kolom-kolom yang tidak memberikan kontribusi yang signifikan terhadap analisis kita. Oleh karena itu, kita bisa mengabaikan atau menghapusnya agar data kita lebih bersih dan terorganisir, serta menghindari kebingungan yang disebabkan oleh informasi yang tidak diperlukan.

````python
drop_col = ['instant', 'temp', 'atemp', 'hum', 'windspeed']
for i in day_df.columns:
  if i in drop_col:
    day_df.drop(labels=i, axis=1, inplace=True)

day_df.head()
````

## Visualization & Explanatory Analysis

### Pertanyaan 1:

Melihat gambaran jumlah pengguna dalam kurun waktu 6 bulan, apakah terdapat sebuah pola tren yang dapat disimpulkan? jika benar adanya pola, bagaimana polanya serta pada tahun berapa serta pada bulan apa jumlah pengguna berapa pada titik tertingginya?

````python
# Kelompokkan data berdasarkan bulan dan hitung jumlah pengguna
monthly_counts = day_df.groupby("mnth").agg({
    "cnt": "sum"
}).reset_index()

# Plot data sebagai grafik bar
plt.figure(figsize=(10, 6))
sns.barplot(
    data=monthly_counts,
    x="mnth",
    y="cnt",
)
plt.title("Jumlah Total Penyewa Sepeda Tahun 2011")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Penyewa")
plt.grid(axis='y')  # Tambahkan grid horizontal untuk mempermudah pembacaan
plt.tight_layout()
plt.show()

# Temukan bulan dengan jumlah pengguna tertinggi
max_users = monthly_counts.loc[monthly_counts['cnt'].idxmax()]
print(f"Jumlah pengguna tertinggi terjadi pada bulan {max_users['mnth']} "
      f"dengan total {max_users['cnt']} penyewa sepeda.")
````

***Pola Tren:***

Dari grafik, kita dapat melihat perubahan jumlah penyewa sepeda selama enam bulan pada tahun 2011:

- Pola Kenaikan: Terdapat tren peningkatan jumlah penyewa dari Januari hingga Juni 2011. Januari dimulai dengan jumlah yang rendah, sekitar 100.000 pengguna. Maret menunjukkan kenaikan signifikan menjadi sekitar 250.000 pengguna. Kenaikan terus terjadi hingga mencapai puncaknya pada bulan Juni.

- Puncak Pengguna:
Jumlah penyewa sepeda tertinggi terjadi pada bulan Juni 2011 dengan total 300.000 lebih pengguna.

### Pertanyaan 2:

Dari analisis, terlihat bahwa jumlah pengguna sepeda menunjukkan tren kenaikan yang konsisten dari bulan Januari hingga Juni, dengan bulan Juni mencatat jumlah tertinggi. Apakah peningkatan jumlah pengguna pada **bulan-bulan yang lebih hangat (April, Mei, dan Juni)** mencerminkan pengaruh cuaca baik terhadap minat masyarakat untuk bersepeda?

````python
# Mengelompokkan data berdasarkan musim dan kondisi cuaca, lalu menghitung rata-rata pengguna
season_weather = day_df.groupby(['season', 'weathersit']).agg({
    'cnt': 'mean'
}).reset_index()

# Plot grafik bar
plt.figure(figsize=(12, 6))
sns.barplot(
    data=season_weather,
    x='weathersit',
    y='cnt',
    hue='season',
    palette='Set2'
)
plt.title('Rata-Rata Jumlah Pengguna Sepeda Berdasarkan Kondisi Cuaca dan Musim')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Rata-Rata Jumlah Pengguna Sepeda')
plt.legend(title='Musim', loc='upper right')
plt.tight_layout()
plt.show()

# Temukan kombinasi musim dan kondisi cuaca dengan jumlah pengguna tertinggi
max_usage = season_weather.loc[season_weather['cnt'].idxmax()]
print(f"Kombinasi musim dan cuaca dengan rata-rata pengguna tertinggi adalah:\n"
      f"Musim: {max_usage['season']}, Kondisi Cuaca: {max_usage['weathersit']}, "
      f"dengan rata-rata pengguna {max_usage['cnt']:.2f}.")
````

**Pola Tren:**

Dari grafik dan keterangan yang diberikan, berikut analisisnya:

- Hubungan Cuaca dan Musim dengan Jumlah Pengguna:
Kombinasi musim dan cuaca dengan rata-rata jumlah pengguna tertinggi adalah Musim Gugur dengan Kondisi Cerah. Kondisi cuaca cerah di semua musim menunjukkan
angka yang lebih tinggi dibandingkan dengan kondisi berawan atau hujan/salju ringan.

- Pola Kenaikan Berdasarkan Cuaca dan Musim:
Bulan April, Mei, dan Juni menunjukkan peningkatan jumlah pengguna sepeda. Ini mungkin berkorelasi dengan musim yang lebih hangat (Musim Panas atau Musim Semi), yang cenderung memiliki kondisi cerah lebih sering. Cuaca cerah memberikan kenyamanan dan motivasi lebih bagi masyarakat untuk melakukan aktivitas di luar ruangan seperti bersepeda.
